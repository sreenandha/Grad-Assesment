##EDA

```python
upos_count = {}
deprel_count = {}
```

```python
for i in train_data:
  if isinstance(i['upos'], list):
    for j in i['upos']:
      if j in upos_count:
        upos_count[j] += 1
      else:
        upos_count[j] = 1

  if isinstance(i['deprel'], list):
    for j in i['deprel']:
      if j in deprel_count:
        deprel_count[j] += 1
      else:
        deprel_count[j] = 1

print(upos_count)
print(deprel_count)
```

```python
for i in upos_count:
  print(i, upos_count[i])
```

```python
for i in deprel_count:
  print(i, deprel_count[i])
```
